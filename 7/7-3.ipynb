```python
import numpy as np
```

```python
X = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
X
```

```python
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
y
```

```python
B1=0.1
```

```python
B0=0.1
```

```python
alpha=1e-6
alpha
```

```python
n=10
```

```python
def mse_(В0, B1, y=y, X=X, n=10): #с интерсептом
    return np.sum((B0+B1*X-y)**2)/n
```

```python
for i in range (3500000):
    y_pred=B0+B1*X
    B0-=alpha*(2/n)*np.sum((y_pred-y))
    B1-=alpha*(2/n)*np.sum((y_pred-y)*X)
    if i%300000==0:
        print('interation: {i}, B0={B0}, B1={B1}, mse={mse}'.format(i=i, B0=B0, B1=B1, mse=mse_(B0,B1)))
```
